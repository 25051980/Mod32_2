# Probabilidade - Tarefa

```python
import numpy as np
from scipy import stats
from scipy.stats import norm
from scipy.stats import norm, t
import matplotlib.pyplot as plt
```

### 1
Uma máquina embala ração para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatório, com distribuição Normal, com média 15kg, mas com desvio padrão 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produção tenha mais que 15,2 kg.

### 1.2
O código calcula a probabilidade de um pacote de ração pesar mais de 15,2 kg, considerando que os pesos seguem uma distribuição normal com média de 15 kg e desvio padrão de 0,1 kg. Utiliza a função de densidade acumulada (`norm.cdf`) para obter a probabilidade acumulada até 15,2 kg e, em seguida, calcula a probabilidade complementar (maior que 15,2 kg) subtraindo este valor de 1. O resultado é exibido em formato decimal.


```python
# Parâmetros da distribuição normal
media = 15  # Média (mu)
desvio_padrao = 0.1  # Desvio padrão (sigma)

# Peso limite para calcular a probabilidade
peso_limite = 15.2

# Calcula a probabilidade acumulada até o peso limite (CDF)
probabilidade_acumulada = norm.cdf(peso_limite, loc=media, scale=desvio_padrao)

# Probabilidade de ser maior que 15,2 kg
probabilidade_maior = 1 - probabilidade_acumulada

# Exibe o resultado
print(f"A probabilidade de que um pacote pese mais que 15,2 kg é: {probabilidade_maior:.4f}")
```

### 2
Da máquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

### 2.1
O código calcula o percentil de 95% para o peso dos pacotes de ração, assumindo que os pesos seguem uma distribuição normal com média de 15 kg e desvio padrão de 0,1 kg. Utiliza a função `norm.ppf`, que representa a inversa da CDF (função de densidade acumulada), para determinar o peso correspondente ao qual 95% dos pacotes têm um peso menor ou igual. O resultado é exibido em quilogramas.


```python
# Parâmetros da distribuição normal
media = 15  # Média (mu)
desvio_padrao = 0.1  # Desvio padrão (sigma)

# Percentil desejado (95%)
percentil = 0.95

# Calcular o valor do percentil (inversa da CDF)
valor_percentil_95 = norm.ppf(percentil, loc=media, scale=desvio_padrao)

# Exibir o resultado
print(f"O percentil de 95% para o peso dos pacotes é: {valor_percentil_95:.4f} kg")
```

### 3
Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na célula abaixo:

### 3.1
O código calcula e visualiza graficamente as probabilidades acumuladas de duas variáveis aleatórias \( Y_1 \) e \( Y_2 \). A variável \( Y_1 \) segue uma distribuição Normal Padrão (média 0 e desvio padrão 1), enquanto \( Y_2 \) segue uma distribuição t-Student com 5 graus de liberdade. As funções `norm.cdf` e `t.cdf` são usadas para calcular as probabilidades acumuladas em um intervalo de valores de \( Y \) entre -6 e 6. O gráfico compara as distribuições, destacando as diferenças nas probabilidades acumuladas, especialmente nas caudas devido às características distintas das distribuições.

```python
# Definir o vetor Y
Y = np.linspace(-6, 6, 200)

# Calcular a probabilidade acumulada para cada valor de Y
# Para a distribuição Normal Padrão
prob_Y1 = norm.cdf(Y, loc=0, scale=1)

# Para a distribuição t-Student com 5 graus de liberdade
prob_Y2 = t.cdf(Y, df=5)

# Plotar os resultados
plt.figure(figsize=(8, 6))
plt.plot(Y, prob_Y1, label="Normal Padrão (Y1)", color="blue")
plt.plot(Y, prob_Y2, label="t-Student (5 GL, Y2)", color="orange")
plt.title("Probabilidade acumulada das variáveis Y1 e Y2")
plt.xlabel("Valores de Y")
plt.ylabel("P(X ≤ Y)")
plt.legend()
plt.grid()
plt.show()
```

O gráfico compara as probabilidades acumuladas de duas distribuições: Normal Padrão (\(Y_1\)) e t-Student com 5 graus de liberdade (\(Y_2\)). Ambas seguem um comportamento semelhante próximo à média, mas diferem nas extremidades (caudas). A distribuição t-Student (\(Y_2\)) apresenta caudas mais largas, refletindo maior probabilidade acumulada para valores extremos, devido à menor quantidade de graus de liberdade. Isso é esperado, pois distribuições t-Student são mais dispersas em relação à Normal.


### 4
Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

### 4.1
O código calcula os percentis para duas distribuições: Normal Padrão (\(Y_1\)) e t-Student com 5 graus de liberdade (\(Y_2\)). Utiliza o vetor \(P\), que contém 20 valores uniformemente espaçados entre 0 e 1, representando probabilidades acumuladas. Para \(Y_1\), a função `norm.ppf` calcula os percentis correspondentes, enquanto para \(Y_2\), utiliza-se a função `t.ppf`. O gráfico resultante mostra como os percentis variam para ambas as distribuições, destacando que a t-Student possui caudas mais largas, refletindo maior dispersão em relação à Normal Padrão.


```python
Y=np.linspace(0, 1, 20)
# Seu código começa aqui
# Definir o vetor P
P = np.linspace(0, 1, 20)  # 20 valores entre 0 e 1

# Calcular os percentis para cada valor de P
# Para a distribuição Normal Padrão
percentis_Y1 = norm.ppf(P, loc=0, scale=1)

# Para a distribuição t-Student com 5 graus de liberdade
percentis_Y2 = t.ppf(P, df=5)

# Exibir os resultados
print("Percentis para a Distribuição Normal Padrão (Y1):")
print(percentis_Y1)

print("\nPercentis para a Distribuição t-Student (Y2):")
print(percentis_Y2)

# Visualizar graficamente os percentis
plt.figure(figsize=(8, 6))
plt.plot(P, percentis_Y1, label="Normal Padrão (Y1)", color="blue", linestyle='-')
plt.plot(P, percentis_Y2, label="t-Student (5 GL, Y2)", color="orange", linestyle='--')
plt.title("Percentis das Distribuições Normal e t-Student")
plt.xlabel("Probabilidade acumulada (P)")
plt.ylabel("Valor Percentil")
plt.legend()
plt.grid()
plt.show()
```

O gráfico compara os percentis das distribuições Normal Padrão (\(Y_1\)) e t-Student com 5 graus de liberdade (\(Y_2\)) em função da probabilidade acumulada (\(P\)). Ambas as distribuições apresentam comportamento semelhante próximo à mediana (\(P=0.5\)), mas nas extremidades, a t-Student (linha laranja) apresenta valores mais extremos devido às suas caudas mais largas. Isso reflete a maior variabilidade da t-Student em comparação com a Normal Padrão, especialmente para valores acumulados próximos de 0 ou 1.
